```python
from pathlib import Path
import pandas as pd
import json

ampere8_path = Path('~/GitHub/FileTransfer/ROSCOE_ampere8').expanduser()
ampere8_perp_path = Path('/lfs/ampere1/0/kaif/GitHub/FileTransfer/ROSCOE_ampere8/perp').expanduser()
ampere1_path = Path('~/GitHub/gold-ai-olympiad/data/MATH/Predictions/ROSCOE').expanduser()
ampere8_comet_path = Path('~/GitHub/FileTransfer/COMET_ampere8/comet_final.pkl').expanduser()
```

```python
def name_map(s):
    if 'NLI_MODEL_SCORES' in s:
        return s[len('NLI_MODEL_SCORES_'):]
    else:
        return s

ampere1_all = {
    name_map(f.stem): json.load(f.open())
    for f in ampere1_path.iterdir()
}
```

```python
ALL_DATASETS = {
    'Algebra_test',
    'Algebra_train',
    'Counting_and_Probability_test',
    'Counting_and_Probability_train',
    'Geometry_test',
    'Geometry_train',
    'Intermediate_Algebra_test',
    'Intermediate_Algebra_train',
    'Number_Theory_test',
    'Number_Theory_train',
    'Prealgebra_test',
    'Prealgebra_train',
    'Precalculus_test',
    'Precalculus_train'
}
```

```python
GRAMMAR_MODEL_SCORES = {
    # f is of the form GRAMMAR_MODEL_SCORES_Prealgebra_test.json
    f.stem[len('GRAMMAR_MODEL_SCORES_'):]: json.load(f.open())
    for f in ampere8_path.glob('GRAMMAR_MODEL_SCORES*.json')
}

num_nans = 0
assert set(GRAMMAR_MODEL_SCORES.keys()) == ALL_DATASETS
for dataset in ALL_DATASETS:
    for i, score in enumerate(GRAMMAR_MODEL_SCORES[dataset]):
        assert 'dataset' in score
        assert 'model' in score
        assert 'i' in score
        assert 'grammar_step' in score
        assert 'grammar_step_max' in score
        for v in score.values():
            if v is None:
                num_nans += 1
                print(score['model'], score['dataset'])
                
print(f'Number of NaNs: {num_nans}')
```

```python
EMB_MODEL_SCORES = {
    # f is of the form EMB_MODEL_SCORES_Prealgebra_test.json
    f.stem[len('EMB_MODEL_SCORES_'):]: json.load(f.open())
    for f in ampere1_path.glob('EMB_MODEL_SCORES*.json')
}

num_nans = 0
assert set(EMB_MODEL_SCORES.keys()) == ALL_DATASETS
for dataset in ALL_DATASETS:
    for i, score in enumerate(EMB_MODEL_SCORES[dataset]):
        assert 'dataset' in score
        assert 'model' in score
        assert 'i' in score
        assert 'faithfulness' in score
        assert 'repetition_step' in score
        assert 'informativeness_step' in score
        assert 'informativeness_chain' in score
        assert 'reasoning_alignment' in score
        assert 'external_hallucination' in score
        assert 'redundancy' in score
        assert 'common_sense_error' in score
        assert 'missing_step' in score
        assert 'semantic_coverage_step' in score
        assert 'semantic_coverage_chain' in score
        for v in score.values():
            if v is None:
                num_nans += 1
                print(score['model'], score['dataset'])
                
print(f'Number of NaNs: {num_nans}')
```

```python
NLI_MODEL_SCORES_ampere8 = {
    # f is of the form NLI_MODEL_SCORES_Prealgebra_test.json
    f.stem[len('NLI_MODEL_SCORES_'):]: json.load(f.open())
    for f in ampere8_path.glob('NLI_MODEL_SCORES_*.json')
}

NLI_MODEL_SCORES_ampere1 = {
    d: scores
    for d, scores in ampere1_all.items() if 'discourse_representation' in scores[0]
}

# assert set(NLI_MODEL_SCORES.keys()) == ALL_DATASETS
print(NLI_MODEL_SCORES_ampere8.keys())
for dataset in NLI_MODEL_SCORES_ampere8:
    for i, score in enumerate(NLI_MODEL_SCORES_ampere8[dataset]):
        assert 'dataset' in score
        assert 'model' in score
        assert 'i' in score
        assert 'discourse_representation' in score
        assert 'coherence_step_vs_step' in score

print(NLI_MODEL_SCORES_ampere1.keys())
for dataset in NLI_MODEL_SCORES_ampere1:
    for i, score in enumerate(NLI_MODEL_SCORES_ampere1[dataset]):
        assert 'dataset' in score
        assert 'model' in score
        assert 'i' in score
        assert 'discourse_representation' in score
        assert 'coherence_step_vs_step' in score
        
NLI_MODEL_SCORES = {
    **NLI_MODEL_SCORES_ampere8,
    **NLI_MODEL_SCORES_ampere1
}
print(f'Number of datasets: {len(NLI_MODEL_SCORES)}')

assert set(NLI_MODEL_SCORES.keys()) == ALL_DATASETS
```

```python
num_nans = 0
for dataset in ALL_DATASETS:
    for i, score in enumerate(NLI_MODEL_SCORES[dataset]):
        assert 'dataset' in score
        assert 'model' in score
        assert 'i' in score
        assert 'discourse_representation' in score
        assert 'coherence_step_vs_step' in score
        for v in score.values():
            if v is None:
                num_nans += 1
                print(score['model'], score['dataset'])

print(f'Number of NaNs: {num_nans}')
```

```python
LANGUAGE_MODEL_SCORES = {
    f.stem: json.load(f.open())
    for f in ampere8_perp_path.iterdir()
}

num_nan = 0
assert set(LANGUAGE_MODEL_SCORES.keys()) == ALL_DATASETS
for dataset in ALL_DATASETS:
    for i, score in enumerate(LANGUAGE_MODEL_SCORES[dataset]):
        for v in score.values():
            if v is None:
                print(score['model'])
                num_nan += 1
        assert 'dataset' in score
        assert 'model' in score
        assert 'i' in score
        assert 'perplexity_step' in score
        assert 'perplexity_chain' in score
        assert 'perplexity_step_max' in score
        
print(f'Number of NaNs: {num_nan}')
```

```python
# save all the scores
save_path = Path('~/GitHub/gold-ai-olympiad/data/MATH/Predictions/ROSCOE_consolidated').expanduser()
save_path.mkdir(exist_ok=True)
for dataset in ALL_DATASETS:
    with open(save_path / f'{dataset}_GRAMMAR_MODEL_SCORES.json', 'w') as f:
        json.dump(GRAMMAR_MODEL_SCORES[dataset], f)
    with open(save_path / f'{dataset}_EMB_MODEL_SCORES.json', 'w') as f:
        json.dump(EMB_MODEL_SCORES[dataset], f)
    with open(save_path / f'{dataset}_NLI_MODEL_SCORES.json', 'w') as f:
        json.dump(NLI_MODEL_SCORES[dataset], f)
    with open(save_path / f'{dataset}_LANGUAGE_MODEL_SCORES.json', 'w') as f:
        json.dump(LANGUAGE_MODEL_SCORES[dataset], f)
```
